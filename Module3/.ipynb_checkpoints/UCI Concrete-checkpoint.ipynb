```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
```

```python
%matplotlib inline
```

```python
matplotlib.style.use('ggplot')
```

```python
concrete = pd.read_excel('Concrete_Data.xls')
```

```python
concrete.head()
```

```python
concrete.columns = ['cement', 'slag', 'ash', 'water', 'plasticizer', 'course_aggregate', 'fine_aggregate', 'age', 'strength']
```

```python
concrete.head()
```

```python
concrete.plot.scatter(x='cement', y='strength')
plt.suptitle('Cement content versus strength')

concrete.plot.scatter(x='water', y='strength')
plt.suptitle('Water content versus strength')

concrete.plot.scatter(x='course_aggregate', y='strength')
plt.suptitle('Water content versus strength')
plt.xlabel('Course Aggregate')
plt.ylabel('Strength')
```

```python
concrete.to_csv('uci_concrete.csv', index=None)
```
