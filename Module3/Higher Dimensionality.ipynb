# Parallel Coordinates

```python
from sklearn.datasets import load_iris

# Import parallel_coordinates
from pandas.tools.plotting import parallel_coordinates

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

# Look pretty...
matplotlib.style.use('ggplot')
# If the above line throws an error, use plt.style.use('ggplot') instead

# Load up SKLearn's Iris Dataset into a Pandas Dataframe
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names) 

df['target_names'] = [data.target_names[i] for i in data.target]

# Parallel Coordinates Start Here:
plt.figure(figsize=(14,8))
parallel_coordinates(df, 'target_names')
plt.show()
```

# Andrew's Scale

```python
from sklearn.datasets import load_iris

## Here import andrews_curves
from pandas.tools.plotting import andrews_curves

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

# Look pretty...
matplotlib.style.use('ggplot')
# If the above line throws an error, use plt.style.use('ggplot') instead

# Load up SKLearn's Iris Dataset into a Pandas Dataframe
data = load_iris()
iris_data = pd.DataFrame(data.data, columns=data.feature_names)
iris_data['target_names'] = [data.target_names[i] for i in data.target]

# Andrews Curves Start Here:
plt.figure(figsize=(14,8))
andrews_curves(iris_data, 'target_names')
plt.show()
```

# Using `imshow`

```python
df = pd.DataFrame(np.random.randn(1000, 5), columns=['a', 'b', 'c', 'd', 'e'])
df.corr()
```

```python
plt.figure(figsize=(14, 8))
plt.imshow(df.corr(), cmap=plt.cm.Blues, interpolation='nearest')

plt.colorbar()
tick_marks = [i for i in range(len(df.columns))]
plt.xticks(tick_marks, df.columns, rotation='vertical')
plt.yticks(tick_marks, df.columns)
```
